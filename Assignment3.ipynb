Task 2: Search Query Keyword Analysis

```python
import string
query = "Buy mobile phone buy phone online"

query = query.lower()
query = query.translate(str.maketrans("", "", string.punctuation))
words = query.split()

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

result = {word: count for word, count in freq.items() if count > 1}

print(result)

```

 Task 4: Email Domain Usage Analysis


```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

domains=[email.split('@')[1] for email in emails]
domain_count = {}
for domain in domains:
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_users = len(emails)
domain_percentage = {domain: (count * 100) // total_users for domain, count in domain_count.items()}



for domain, percentage in domain_percentage.items():
    print(f"{domain}: {percentage}%")

```

 Task 5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
spikes = [(day, value) for day, value in enumerate(sales, start=1) if value > average_sales * 1.3]
print(f"Average Daily Sales: {average_sales:.2f}")
print("Sales Spikes (Day, Value):")
print(spikes)
```

Task 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
freq = {}
for uid in user_ids:
    freq[uid] = freq.get(uid, 0) + 1
duplicates = {uid: count for uid, count in freq.items() if count > 1}
for uid, count in duplicates.items():
    print(f"{uid} → {count} times")
```
