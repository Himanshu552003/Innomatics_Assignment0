Task-2 Pass/Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]

pass_count=0
fail_count=0

for mark in marks:
    if mark >=50:
        pass_count +=1
    else:
        fail_count+=1

print("Pass Student: ", pass_count)
print("Fail Student: ", fail_count)
```

Task-3 : Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
for i in range(len(names)):
    names[i]=names[i].strip().lower()

print('Cleaned Name: ', names)
```

Task-4 : Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length=len(msg)
    print(f"Message: '{msg}'")
    print(f"Length: {length}")
    if length>10:
        print("Flagged (too long)")
    else:
        print("")
```
