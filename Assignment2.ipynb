Task 1 : Unique Words in a Sentence


```python
sentence= "python is easy and python is powerful"

words=sentence.split()

unique_words=[]
for word in words:
    if word != unique_words:
        unique_words.append(word)

print("Unique word count:", len(unique_words))
print("Unique words:", set(unique_words))
```

Task 2: Highest Salary from Employee Data

```python
details= {"Ravi":7500, "Anita":68000, "Kiran":72000}
highest_sal=0
highest_employee= ""

for name in details:
    if details[name]>highest_sal:
        highest_sal=details[name]
        highest_employee=name
print("Highest Salary:", highest_employee, "-", highest_sal)
```

Task 3: Find Maximum and Minimum Values

```python
numbers= [45,22,89,10,66]
max=numbers[0]
min=numbers[0]

for num in numbers:
    if num>max:
        max=num
    if num<min:
        min=num

print("Max:", max)
print("Min: ", min)
```

Task 4: Count Products Above a Price Threshold


```python
prices = [450, 1200, 899, 1500, 300]
count=0

for amt in prices:
    if amt>1000:
        count+=1

print("Product above 1000:", count)       
```

Task 5: Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]
total=len(attendance)
present=0
for day in attendance:
    if day=="P":
        present+=1


attendance_percentage=(present/total)*100

print("Attendance Percentage:", attendance_percentage)
```

Task 7: Count Character Frequency


```python
text='python'
count={}
for ch in text:
    if ch in count:
        count[ch]+=1
    else:
        count[ch]=1
print(count)
```
